## Additional Helper Functions

````python
def hsv_image(image):
    #Place holder for future development. 
    # this color space for robustness against variations in illumination.
    return cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

def yCrCb_image(image):
    #Place holder for future development. I would like to explore 
    # this color space for robustness against variations in illumination.
    return cv2.cvtColor(image, cv2.COLOR_RGB2YCrCb)

def rgb_filter(image):
    #Calculate image size
    xsize = image.shape[1]
    ysize = image.shape[0]
    #Create image copy
    color_select = np.copy(image)

    #Define color selection thresholds
    r_thres = 175
    g_thres = 155
    b_thres = 5
    rgb_thres = [r_thres, g_thres, b_thres]
    
    # Identify pixels below the threshold
    thresholds = (image[:,:,0] < rgb_thres[0]) \
               | (image[:,:,1] < rgb_thres[1]) \
               | (image[:,:,2] < rgb_thres[2])
    color_select[thresholds] = [0,0,0]
    return color_select
````

## Build a Lane Finding Pipeline



````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    #return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    rm = [] # right lane line slope list
    lm = [] # left  lane line slope list
    rc = [] # right lane line centers list
    lc = [] # left  lane line centers list
    
    #iterate through each line points (x1,y1,x2,y2)
    # calculate slope
    for line in lines:
        x1 = line[0][0]
        y1 = line[0][1]
        x2 = line[0][2]
        y2 = line[0][3]
        slope = (y2 - y1)/(x2 - x1)
        center = [(x2 + x1)/2, (y2 + y1)/2]

        #reject slopes that are outliers (i.e., too horizontal or too verticla), else add to list of slopes and centers
        if (slope > 0):
            if 0.5 < slope < 0.8:
                rm.append(slope)
                rc.append(center)
        elif (slope < 0):
            if -0.8 < slope < -0.5:
                lm.append(slope)
                lc.append(center)
            
    y_upp = 350 # one-half the image height
    y_bot = 535 # image height

    ###Find average slope and center points
    #Line on left side of image
    if len(lm) != 0:
        #Calculate average slope
        l_slope=np.sum(lm)/len(lm)
        #Calculate average center point
        l_center=np.divide(np.sum(lc,axis=0),len(lc))
        #Find x-coordinates of extrapolated endpoints (x1_l=near bottom of screen)        
        x1_l = int( (y_bot - l_center[1])/l_slope + l_center[0] )
        x2_l = int( (y_upp - l_center[1])/l_slope + l_center[0] )
        #Draw left line on input image (mutates in place!)
        cv2.line(img, (x1_l, y_bot), (x2_l, y_upp), color, thickness)
        
    #Line on right side of image
    if len(rm) != 0:
        #Calculate average slope
        r_slope=np.sum(rm)/len(rm)
        #Calculate average center point
        r_center=np.divide(np.sum(rc,axis=0),len(rc))
        #Find x-coordinates of extrapolated endpoints (x1_r=near bottom of screen)        
        x1_r = int( (y_bot - r_center[1])/r_slope + r_center[0] )
        x2_r = int( (y_upp - r_center[1])/r_slope + r_center[0] )
        #Collect line endpoints into an np.array
        #averaged_lines = np.array( [x1_r, y_bot, x2_r, y_upp, x1_l, y_bot, x2_l, y_upp] )
        #Draw right line on input image (mutates in place!)
        cv2.line(img, (x1_r, y_bot), (x2_r, y_upp), color, thickness)
        
   
    
def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    # lines is an (nx1x4) element np.array with x,y coordinates for each hough line endpoints 
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    # draw_lines returns a black background with hough lines drawn on it
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
````

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images directory.
#image = mpimg.imread('test_images/whiteCarLaneSwitch.jpg')
def image_pipeline(image):
    #White lane stripe defined by Federal Standard Color #595-17886
    #white_RGB = [247, 241, 227]
    #Yellow lane stripe defined by Federal Standard Color #595-33538
    #white_RGB = [255, 183, 0]
    
    #Convert to yCrCb color space
    #r1 = yCrCb_image(image)
    #plt.imshow(r1)    
    
    #Apply color filter
    color_select = rgb_filter(image)
    
    #Grayscale conversion
    gray = grayscale(color_select)
    #plt.imshow(gray)

    #Define a kernel size and apply Gaussian smoothing
    kernel_size = 3
    blur_gray = gaussian_blur(gray, kernel_size)
    #plt.imshow(blur_gray)

    #Define our parameters for Canny and apply
    low_threshold = 50
    high_threshold = 150
    edges = canny(blur_gray, low_threshold, high_threshold)
    #plt.imshow(edges) 

    #Calculate image size
    xsize = image.shape[1]
    ysize = image.shape[0]
    #ROI mask
    vertices = np.array([[(0.04*xsize, 0.99*ysize), (0.48*xsize, 0.60*ysize), 
                          (0.52*xsize, 0.60*ysize), (0.96*xsize, 0.99*ysize)]], dtype=np.int32)
    #print(vertices)
    masked_edges = region_of_interest(edges, vertices)   

    #Hough transform 
    rho = 2 # distance resolution in pixels of the Hough grid
    theta = np.pi/180    # angular resolution in radians of the Hough grid
    threshold = 40       # minimum number of votes (intersections in Hough grid cell)
    min_line_len = 5    # minimum number of pixels making up a line
    max_line_gap = 300    # maximum gap in pixels between connectable line segments
    raw_lines = hough_lines(masked_edges, rho, theta, threshold, min_line_len, max_line_gap)
    #plt.imshow(raw_lines)
    
    #Draw averaged lines on color image
    weighted_lines = weighted_img(raw_lines, image, α=0.8, β=1., λ=0.)
    #plt.imshow(weighted_lines)
   
    return weighted_lines
````
